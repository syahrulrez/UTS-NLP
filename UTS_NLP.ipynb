3. NORMALIZATION

```python
data = df['Data'].dropna().iloc[0]

data = re.sub(r"\d+", "", data)
data = data.translate(str.maketrans('', '', string.punctuation))
data = re.sub(r'\s+', ' ', data).strip()

tokens = word_tokenize(data)

kamus_normalisasi = {
    'yg'  : 'yang',
    'dgn' : 'dengan',
    'utk' : 'untuk',
    'tdk' : 'tidak',
    'krn' : 'karena',
    'jg'  : 'juga',
    'sdh' : 'sudah',
    'blm' : 'belum',
    'dpt' : 'dapat',
    'dlm' : 'dalam',
    'tsb' : 'tersebut',
}

hasil = [kamus_normalisasi.get(kata, kata) for kata in tokens]

print(hasil)
```

4.  STOPWORD REMOVAL

```python
import re
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd

data = df['Data'].dropna().iloc[0]

data = re.sub(r"\d+", "", data)
data = data.translate(str.maketrans('', '', string.punctuation))
data = re.sub(r'\s+', ' ', data).strip()

tokens = word_tokenize(data)

nltk.download('stopwords')
listStopword = set(stopwords.words('indonesian'))

removed = []
for t in tokens:
    if t not in listStopword:
        removed.append(t)

print(removed)
```

7. LEMMATIZATION

```python
import re
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
import string
import nltk

nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')

data = df['Data'].dropna().iloc[0]

data = re.sub(r"\d+", "", data)
data = data.translate(str.maketrans('', '', string.punctuation))
data = re.sub(r'\s+', ' ', data).strip()
data = data.lower()

tokens = word_tokenize(data)
tagged_tokens = pos_tag(tokens)

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'

lemmatizer_sentence = []
for word, tag in tagged_tokens:
    if word.lower() == 'are' or word.lower() in ('is', 'an'):
        lemmatizer_sentence.append(word.lower())
    else:
        lemmatizer_sentence.append(
            lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        )

print("Original Sentence: ", data)
print("Lemmatized sentence:", " ".join(lemmatizer_sentence))
```
